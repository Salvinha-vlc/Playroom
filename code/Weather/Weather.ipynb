This dataset is a small subset of a global historical climatology network dataset. The data represents the daily weather records for a weather station in Mexico for the whole 2010 (except September for which no data was stored). The goal of this notebook is to get a tidy dataset able that conveys information at first sight.

```python
import pandas as pd
```

```python
weather = pd.read_csv("Data/temperatures.csv")
```

```python
weather.head()
```

```python
weather = weather.iloc[:,3:]
```

```python
weather.head()
```

```python
len(weather)
```

```python
weather
```

```python
weather_melted = pd.melt(frame=weather, id_vars=["month", "element"])
```

```python
weather_melted.head()
```

```python
weather_unstacked = weather_melted.groupby(["month", "element"]).value.agg([max, min]).unstack()
```

```python
weather_unstacked.head()
```

```python
weather_clean = weather_unstacked.loc[:,[("max", "tmax"), ("min", "tmin")]]
```

```python
weather_clean.head()
```

```python
weather_clean.columns = weather_clean.columns.droplevel(1)
```
